```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = r'C:\Users\Admin\Desktop\Salary_dataset1.csv'  
df = pd.read_csv(file_path)


x = df['YearsExperience'].values  
y = df['Salary'].values


xmean = sum(x) / len(x)
ymean = sum(y) / len(y)


numerator = sum((xi - xmean) * (yi - ymean) for xi, yi in zip(x, y))
denominator = sum((xi - xmean) ** 2 for xi in x)

theta1 = numerator / denominator
theta0 = ymean - theta1 * xmean

print(f"Slope (theta1): {theta1}")
print(f"Intercept (theta0): {theta0}")


def predict(xvalues):
    return [theta0 + theta1 * xi for xi in xvalues]


xtest = [2, 4, 3.2]
ypred = predict(xtest)

print(f"Predictions for {xtest}: {ypred}")


plt.scatter(x, y, color='blue', label='Original Data')
xline = range(int(min(x)), int(max(x)) + 1)  
yline = predict(xline)  
plt.plot(xline, yline, color='red', label='Regression Line')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.show()
```
